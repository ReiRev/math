2次方程式$z^2+g(k)z+k=0$の2つの解について調べる．これらは，実数$k$を$-\inf < k < \inf$の範囲で動かすと，複素平面上でどのような軌跡を描くか．次の場合について求めよ．

## (1)

$g(k)=1$

## 解法

$z^2+g(k)z+k=0$の解は，$z= (-1\pm \sqrt{g(k)^2 + k})/2$なので，以下．

```python
import sys
sys.path.append('../../../')
from plot import plot_cp
import numpy as np

def g(k):
    return 1

k = np.linspace(-10000, 10000, 10000000)
z_plus = (-1 + np.sqrt(g(k)**2-4*k+0j))
z_minus = (-1 - np.sqrt(g(k)**2-4*k+0j))
z = np.concatenate([z_minus, z_plus])
plot_cp(z, lim=[-8, 8])
```
